```python
import math 

def number_to_list(n):
    order =  2 ** math.floor(math.log(n,2))
    binary_list = []  
    while order >= 1:
        binary_list.append( int(n // order) )
        n %= order
        order /= 2
    return binary_list      
    
    
def list_to_number(l):
    n = 0
    multiplier = 1 
    for z in reversed(l):
        n+= multiplier * z
        multiplier *= 2
    return n


def energy(l):
    last = 0
    total = 0
    for z in l + [0]:
        if z == last:
            continue
        else:
            total += 1
            last = z
    return total//2


```

```python
for n in range(1, 100):
    l = number_to_list(n)
    print('{:3}: {}  E: {}'.format(n, l, energy(l)) )
```
